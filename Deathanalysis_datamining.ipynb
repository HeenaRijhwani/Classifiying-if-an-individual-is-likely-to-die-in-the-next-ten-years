```python
#Data seems to be imbalanced
#Performing SMOTE to resample minority class (class=1)
import imblearn
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X2 = df1[["exercise",'pepticulcer','weakheart','nerves','qsmk','nervousbreak',"hbp",'marital','infection','lackpep','otherpain','wtloss','hbpmed','pica',"hayfever",'polio','sex','tumor','race','tb',"active","headache","hepatitis","hf","education","diabetes","alcoholfreq","alcoholtype","alcoholpy","allergies","asthma","boweltrouble","bronch",'colitis','chroniccough',"alcoholhowmuch",'cholesterol',"dbp",'income','price','ht','wt','sbp','tax',"age","pregnancies",'smokeintensity','smokeyrs']].values
y2 = df1['Class'].values
oversample = SMOTE()
X, y = oversample.fit_resample(X2, y2)

scaler = StandardScaler()
X4=scaler.fit_transform(X) 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_predict

lor = LogisticRegression()
X=X4
y_pred = cross_val_predict(lor, X, y, cv=10)
lor_accuracy=round(accuracy_score(y,y_pred),4)*100
print("Accuracy for LOR: {}%".format(lor_accuracy))
#or (cross_val_score(lor,X,y,cv=10).mean())*100
```

```python
#overall performace metrics
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
cm=confusion_matrix(y,y_pred)
print(cm)
TP=cm[0,0]
print("True positives are", TP)
FN=cm[0,1]
print("False negatives are", FN)
FP=cm[1,0]
print("False positives are", FP)
TN=cm[1,1]
print("True negatives are", TN)
TPR=TP/(TP + FN)
print("True positive rate is", np.round(TPR,2))
print("Recall is", np.round(TPR,2))
TNR=TN/(TN + FP)
print("True negative rate is", np.round(TNR,2))
Precision=TP/(TP+FP)
print("Precision is",Precision)
MCC=((TN*TP)-(FP*FN))/np.sqrt((TP+FP)*(TP+FN)*(TN+FN)*(TN+FP))
print("Matthews correlation coefficient is", MCC)
FPR=FP/(FP + TN)
print("False positive rate is", np.round(FPR,2))
FNR=FN/(FN + TP)
print("False negative rate is", np.round(FNR,2))
F1=(2 * (Precision*TPR)) / (Precision+TPR)
print("F measure is", F1)

#Roc area/ Auc
fpr, tpr, _ = roc_curve(y, y_pred)
roc_auc = auc(fpr,tpr)
print("Roc area is", roc_auc)
```
